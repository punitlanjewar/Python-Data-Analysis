## Order Priority

```python
df['Order Priority'].value_counts()
```

```python
df['Order Priority'].unique()
```

```python
df['Order Priority'].replace('Critical ', 'Critical')
```

```python
pyl.figure(figsize=(5,3))
sns.countplot(x='Order Priority', data=df)
pyl.show()
```

```python
pyl.figure(figsize=(5,5))
pyl.title('Count of Order Priority')
ax=sns.countplot(x='Order Priority', data=df)
for count in ax.containers:
    ax.bar_label(count)
pyl.savefig('Count of Order Priority.jpg')
```

## Ship Mode

```python
df['Ship Mode'].value_counts()
```

```python
x = df['Ship Mode'].value_counts().index
y = df['Ship Mode'].value_counts().values
pyl.figure(figsize=(5,5))
pyl.pie(y,labels=x,autopct='%0.2f%%', startangle=90)
pyl.show()
pyl.savefig('Ship Mode Piechart.jpg')
```

```python
pyl.figure(figsize=(5,5))
ax=sns.countplot(x='Ship Mode', data=df, hue='Product Category')
for count in ax.containers:
    ax.bar_label(count)
pyl.show()
```

#### From above figure we can see that the count of Office Supplies are more using Regular Air shipping and express Air shipping, and in Trusk Delivery the furniture is more.
