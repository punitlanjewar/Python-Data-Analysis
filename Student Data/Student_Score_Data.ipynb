```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.info()
```

```python
df['WklyStudyHours'] = df['WklyStudyHours'].replace('05-Oct', '5-10')
```

```python
df.head(10)
```

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(x='Gender', data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(7,5))
ax = sns.countplot(x='Gender', hue='TransportMeans', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

#### From above we can analyse that the count of female candidates are more those who are travelling by school bus

```python
df.columns
```

```python
gb = df.groupby('ParentEduc').agg({'MathScore':'mean', 'ReadingScore':'mean', 'WritingScore':'mean'})
sns.heatmap(gb, annot=True)
plt.show()
```

#### From above chart we can analyse that parent education is impacting on childrens scores.

```python
gb = df.groupby('ParentMaritalStatus').agg({'MathScore':'mean', 'ReadingScore':'mean', 'WritingScore':'mean'})
sns.heatmap(gb, annot=True)
plt.show()
```

#### From above graph we can tell that parents marital status won't much affecting the scores.
