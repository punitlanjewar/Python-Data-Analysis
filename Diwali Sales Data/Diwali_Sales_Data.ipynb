Data Analysis

Data Analysis By Gender

```python
sns.countplot(x = 'Gender', data = df)
```

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
visual = sns.barplot(x = 'Gender', y = 'Amount', data = sales_gender)
for bars in visual.containers:
    visual.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

From above data and figure we will seee that the most of the buyers are female customers, the ratio of female customers are greater than male customers.

```python
df.columns
```

```python
# Orders from top 10 States
sale_state = df.groupby(['State', 'Gender'], as_index=False)['Orders'].sum().sort_values(by=['Orders'], ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Orders', hue='Gender',data=sale_state)
```

```python
# total amount of sale in 10 states
amount_state = df.groupby(['State', 'Gender'], as_index=False)['Amount'].sum().sort_values(by=['Amount'], ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x='State', y='Amount', hue='Gender', data=amount_state)
```
